```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# S&P 500

```python
H_1=pd.read_excel("H=1.xlsx",index_col='Date',parse_dates=True,sheet_name="S&P500")
H_3=pd.read_excel("H=3.xlsx",index_col='Date',parse_dates=True,sheet_name="S&P500")
H_5=pd.read_excel("H=5.xlsx",index_col='Date',parse_dates=True,sheet_name="S&P500")
H_10=pd.read_excel("H=10.xlsx",index_col='Date',parse_dates=True,sheet_name="S&P500")
H_22=pd.read_excel("H=22.xlsx",index_col='Date',parse_dates=True,sheet_name="S&P500")
```

```python
H_1.reset_index(inplace=True)
H_3.reset_index(inplace=True)
H_5.reset_index(inplace=True)
H_10.reset_index(inplace=True)
H_22.reset_index(inplace=True)
```

```python
plt.plot(H_1["Date"], H_1["Actuals"])
plt.plot(H_1["Date"], H_1["GARCH"])  
plt.ylabel("S&P 500 Realized Volatility")
```

```python
plt.plot(H_3["Date"], H_3["Actuals"])
plt.plot(H_3["Date"], H_3["GARCH"])  
plt.ylabel("S&P 500 Realized Volatility")
```

```python
plt.plot(H_5["Date"], H_5["Actuals"])
plt.plot(H_5["Date"], H_5["GARCH"])  
plt.ylabel("S&P 500 Realized Volatility")
```

```python
plt.plot(H_10["Date"], H_10["Actuals"])
plt.plot(H_10["Date"], H_10["GARCH"])  
plt.ylabel("S&P 500 Realized Volatility")
```

```python
plt.plot(H_22["Date"], H_22["Actuals"])
plt.plot(H_22["Date"], H_22["GARCH"])  
plt.ylabel("S&P 500 Realized Volatility")
```
