# STOXX50E

```python
df=pd.read_excel("Stocks from other geographics returns.xlsx", sheet_name="STOXX50E",index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
returns=df["Returns"][:int(len(df["Returns"])*0.7)]*100
```

```python
!pip install arch
```

```python
from arch import arch_model
```

```python
for i in range(8):
  for j in range(8):
    p=i+1
    q=j+1
    am = arch_model(returns, vol="Garch", p=i+1, o=0, q=j+1, dist="Normal")
    res = am.fit(update_freq=5)
    print(f"GARCH({p,q})")
```

```python
am = arch_model(returns, vol="Garch", p=1, o=0, q=1, dist="Normal")
res = am.fit(update_freq=5)
forecasts = res.forecast(horizon=1, reindex=False)
print(forecasts.variance.iloc[:])
```

```python
res
```

```python
for i in range(8):
  for j in range(8):
    for k in range(8):
      o=i+1
      p=i+1
      q=j+1
      am = arch_model(returns, p=i+1, o=k+1, q=j+1, dist="Normal")
      res = am.fit(update_freq=5)
      print(f"DJR-GARCH({p,o,q})")
```

```python
am = arch_model(returns, p=1, o=1, q=1, dist="Normal")
res = am.fit(update_freq=5)
```
